```python
import numpy as np

def anchor_reward(response: str, anchor_correct: np.ndarray, anchor_position: np.ndarray, beta: float = 1.5) -> float:
    """
    Aggregates a boolean array (True=correct, False=wrong) to a scalar score between 0 and 1,
    using a weighted average where weights are exp(beta * (anchor_position / response_length)).
    Anchors closer to the end of the response have higher weight.
    """
    if anchor_correct.size == 0:
        return 0.0
    # anchor_position: np.ndarray of int, anchor_correct: np.ndarray of bool
    # anchor_position is the position of each anchor in the response (0-based)
    # response_length is the max position + 1 (assuming positions are valid)
    response_length = len(response)
    length_ratios = anchor_position / response_length
    weights = np.exp(beta * length_ratios)
    weighted_score = np.sum(anchor_correct * weights) / np.sum(weights)
    weighted_score = max((weighted_score - 0.5) / 0.5, 0.0)
    return float(weighted_score)
```

```python
response = "The answer is 100."
anchor_correct = np.array([True, False, True])
anchor_position = np.array([1, 15, 16])
beta = 1.5
anchor_reward(response, anchor_correct, anchor_position, beta)
```
